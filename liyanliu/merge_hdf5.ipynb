```python
import h5py
import os
import numpy as np

currentpath = os.getcwd()
path = currentpath + '/InOut_100X100'
print("currentpath:", currentpath,"path:", path)

def obtainFilesL(path):
    filesL = []
    for f in os.listdir(path):
        filesL.append(f)
    return filesL
filesL = obtainFilesL(path)
    
print("len(filesL):", len(filesL))
#print(filesL),print()


def obtainTimelables(filesL):
    data = '201504'
    timelabels = []
    for d in range(len(filesL)):
    #for d in range(17):
        day = filesL[d][5:7]
        for h in range(24):
            for m in range(1,7):
                if len(str(h))<2:
                    temp = data +day +'0'+str(h)+str(m)+'0'
                else:
                    temp = data +day +str(h)+str(m)+'0'
                timelabels.append(int(temp))
    return timelabels

timelabels = obtainTimelables(filesL)
print("len(timelabels):", len(timelabels))
print(timelabels[:10])
```

```python
dataset = []
for i in range(len(filesL)):
#for i in range(1):
    filePath = path + '/'+ filesL[i]
    #filePath = path + '/HZ14_14_M100x100_InOut.hdf5' 
    print(filePath)
    f = h5py.File(filePath)
    for ke in f.keys():
        print("ke, f[ke].shape:", ke, f[ke].shape)
        dataset.append(f[ke])
```

```python
N =len(filesL)*24*6

inOutTk = np.zeros((N, 2, 100, 100))
print("inOutTk.shape:", inOutTk.shape, "N:", N)

print(len(dataset))
for t in range(len(dataset)):
    inOutTk[t*144:((t+1)*144)] = dataset[t]
print(inOutTk.shape)    

for i in range(len(inOutTk)):
    print(i,sum(sum(inOutTk[i][0])))

f = h5py.File("HZ15_M100x100_InOut.hdf5", "w")
f['data'] = inOutTk
f['labels'] = timelabels
```
