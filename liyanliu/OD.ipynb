```python
import numpy as np
import pandas as pd
import os
import math
import time
import h5py

current_path = os.getcwd()
print("current_path:", current_path),print()

bottom=30.012
up=30.4439     
left=119.8691 
right=120.461
blockCount=50

wideRange=right-left
heightRange=up-bottom
wideDistance=wideRange/blockCount
heightDistance=heightRange/blockCount

data = pd.read_csv("filesList.csv")
listName = data['filename']
print("listName:", type(listName[0]))

def deviceFile():
    fileList = []
    for i in range(24):
        if len(str(i))<2:
            fileList.append('DeviceID'+'0'+str(i))
        else:
            fileList.append('DeviceID'+str(i))
    return fileList

fileList = deviceFile()
print("fileList:", fileList),print()

def deviceId(m,n):
    deviceList = []
    path = current_path + '/'+listName[m]+'/'+fileList[n]+'/'
    for file in os.listdir(path):
        deviceList.append(file)
    return deviceList,path

deviceList,path = deviceId(0,1)
print("len(deviceList):", len(deviceList))
print("path:", path)
print(listName[0])
```

```python
print("开始运行......")
now = time.clock()

def trafficVolum(m):
    TV = np.zeros((24,50,50))
    for n in range(24):
    # for n in range(0,2):
        deviceList,path = deviceId(m,n) 
        print("len(deviceList):", len(deviceList))
        print('path:',path)
        trafficVolume = np.zeros((50,50))
        for i in range(len(deviceList)):
    #     for i in range(0,3):
            data = pd.read_csv(path+deviceList[i])
    #         print("deviceList[i]:", deviceList[i])
            df = pd.DataFrame(data = data)
            rowList = [min(math.floor(row),49) for row in (df.Latitude-bottom)/heightDistance]
            colList = [min(math.floor(col),49) for col in (df.Longitude-left)/wideDistance]
            grid = list(set((str(rowList[i])+','+str(colList[i])) for i in range(len(rowList))))
    #         print(len(grid))
            for j in range(len(grid)):
                row = int(grid[j].split(',')[0])
                col = int(grid[j].split(',')[1])
                trafficVolume[row][col] +=1
    #     trafficVoume = np.array(trafficVolume)
        print(sum(sum(trafficVolume)))
        TV[n]=np.array(trafficVolume)

    f = h5py.File(current_path + '/d01_w4/trafficVolume/'+ listName[m]+'_M50x50_trafficVolume.hdf5', "w")
    print(current_path + '/d01_w4/trafficVolume/'+ listName[m]+'_M50x50_trafficVolume.hdf5')
    f['data'] = TV
    
trafficVolum(17)    
    
print("运行完毕！耗时%f分钟。"%((time.clock()-now)/60))
```

```python
import numpy as np
import pandas as pd
import h5py
import os
current_path = os.getcwd()

data = pd.read_csv("filesList.csv")
listName = data['filename']
print("listName:", type(listName[0]))

# 得到tensor,  location X day X time = trafficVolume,  (2500X7X24)
def obtainTensor(listName):
    dataset = []
    for i in range(7):
        f = h5py.File(current_path + '/d01_w4/trafficVolume/'+ listName[i]+'_M50x50_trafficVolume.hdf5')
        for ke in f.keys():
            dataset.append(f[ke])
    print("len(dataset):", len(dataset))

    num = 0
    for n in range(len(dataset)):
        for j in range(len(dataset[0])):
#             print(n,j,dataset[n][j].shape)
            data = dataset[n][j].reshape(1,2500)[0]
            index = data.nonzero()[0] # 得到data中非零元素的索引
            print(index,len(index))
            num += len(index)
            value = [int(data[i]) for i in index] # 得到data中与索引相对应的非零元素的值
            print(len(value)),print()
    #         print(current_path+'/tensorData/loc_day_time.csv')
            tensorPath = current_path+'/tensorData/loc_day_time.csv'
            with open(tensorPath,'a') as f:
                for i in range(len(index)):
                    f.write(str(index[i])+','+str(n+1)+','+str(j+1)+','+str(value[i])+'\n')
    print("num:",num)

obtainTensor(listName)
```
